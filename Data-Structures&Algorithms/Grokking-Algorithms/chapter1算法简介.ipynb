### 1.1二分查找
二分查找是对于有序列表来说，不是有序的话无法执行，因为二分查找思想每次都要和中间取值比大小来确定范围，无序就确定不了中间值啦。

比较难理解的点：找到最后两个怎么找？由于二分查找，最重要的是移动索引，最后两个数索引相邻，相加÷2肯定为前者索引，先判断前者是不是，不是的话前者索引+1和后者索引相同，那最后一个查找的索引必然是后者，就确定了。

二分查找的复杂度为O(logn)

```python
# 二分查找python实现
def binary_search(list, item):
    low=0
    max=len(list)-1

    while(low <= max):
        mid = int((low + max) / 2)
        value = list[mid]

        if value == item:
            return mid
        elif value > item:
            max = mid - 1
        else:
            low =  mid + 1
    return None

test_list=list(range(1, 101))
research_index = binary_search(test_list, 10)
research_index
        
```

### 1.2 时间复杂度
大O表示法是一种特殊的表示法，指出了算法运行有多快，但是需要特别注意的一点，算法的运行时间是以不同的速度增加的，比如O(n)和O(logn)两者随着n的运行时间变化极速差距变大。

大O表示法记录了最糟糕情况下的操作数


基本的时间复杂度排序O(logn) < O(n) < O(nlogn) < O(n2) < O(n!) 


```python
import matplotlib.pyplot as plt
import numpy as np
from math import factorial

# 定义n的范围
n_values = np.arange(1, 20, 1)  # 避免n!增长过快，取1到20
n_large = np.linspace(1, 100, 500)  # 其他函数适用较大范围

# 定义函数
log_n = np.log(n_large)
n = n_large
n_log_n = n_large * np.log(n_large)
n_square = n_large**2
n_factorial = [factorial(i) for i in n_values]  # 计算小范围内的n!

# 绘制图形
plt.figure(figsize=(10, 6))

# 绘制不同的函数
plt.plot(n_large, log_n, label='$\log n$', color='blue')
plt.plot(n_large, n, label='$n$', color='orange')
plt.plot(n_large, n_log_n, label='$n \log n$', color='green')
plt.plot(n_large, n_square, label='$n^2$', color='red')
plt.plot(n_values, n_factorial, label='$n!$', color='purple', marker='o')  # 使用离散点表示n!

# 添加标题和标签
plt.title("Comparison of Functions: $\log n$, $n$, $n \log n$, $n^2$, $n!$")
plt.xlabel('$n$')
plt.ylabel('Value')
plt.yscale('log')  # 使用对数刻度展示增长趋势
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```
